# IMAGE TO SKETCH PROJECT

This project uses Numpy and OpenCV to convert an image from normal to a sketch verison of itself.


# 1.) Step by step process


*  Image selection upload the image we want to convert

*  Read the image

* Convert it to grayscale

* Invert the grayscale image

* Apply Gaussian blur to the inverted image

* Invert the blurred image

* Blend the grayscale with the inverted blur using cv2.divide()




> We successfully converted the images from normal to sketch form.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read and resize the image (optional)
img = cv2.imread('tiger_sketch.jpg')
img = cv2.resize(img, (500, 500))  # Optional: resize for display

# Step 2: Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 3: Invert the grayscale image
inverted = 255 - gray

# Step 4: Apply Gaussian blur
blurred = cv2.GaussianBlur(inverted, (21, 21), sigmaX=0, sigmaY=0)

# Step 5: Invert the blurred image
inverted_blur = 255 - blurred

# Step 6: Create the pencil sketch using divide
sketch = cv2.divide(gray, inverted_blur, scale=256.0)

# Show results
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sketch, cmap='gray')
plt.title("Pencil Sketch")
plt.axis('off')

plt.tight_layout()
plt.show()
```
